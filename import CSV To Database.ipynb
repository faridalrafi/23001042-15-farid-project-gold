```python
#Import Library
import pandas as pd
import sqlite3 as sq3
```

```python
#Create connection
conn = sq3.connect("database_gold.db")
print("Open database berhasil")
```

```python
df_kamusalay = pd.read_csv("new_kamusalay.csv", names = ['alay', 'normal'], encoding="cp1252")
df_kamusalay.head(10)
```

```python
df_kamusalay.to_sql('kamusalay', conn, if_exists = 'replace', index = False)
df_kamusalay = pd.DataFrame(df_kamusalay)
df_kamusalay
```

```python
df_abusive = pd.read_csv('abusive.csv', encoding="cp1252")
df_abusive.head()
```

```python
df_abusive.to_sql('abusive', conn, if_exists = 'append', index = False)
df_abusive = pd.DataFrame(df_abusive)
df_abusive
```

```python
conn.commit()
conn.close()
```
